# A Short Example

```python
# create some data we want to plot
n = np.array([0,1,2,3,4,5])
x = np.linspace(0, 5, 10)
xx = np.linspace(-0.75, 1., 100)
```

```python
# plot it (edit me)
fig, axes = plt.subplots(1, 4, figsize=(12,3))
axes[0].scatter(xx, xx + 0.75*np.random.random(len(xx)))
axes[1].step(n, n**2, lw=2)
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
```

# An Animated Example

```python
fig = plt.figure(figsize=(12,3)); ax=plt.axes(xlim=(0,2),ylim=(-2,2))
line, = ax.plot([], [], lw=2)
def animate(i, _line):
    x = np.linspace(0,2,1000); y=np.sin(3*np.pi*(x-0.01*i))
    _line.set_data(x, y)
    return _line,
animation.FuncAnimation(fig, animate, fargs=(line,), frames=50, interval=20, blit=True)
```

# Works with Gtk and matplotlib!

```python
def _noise(btn, _ax,_canv):
    _ax.plot(x,np.sin(x*np.random.random()))
    _canv.draw()
    
w = Gtk.Window(default_width=400,default_height=400)
box = Gtk.Box(orientation=Gtk.Orientation.VERTICAL)
fig = Figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.plot(x,np.sin(x))
canvas = FigureCanvasGTK3Agg(fig)
box.pack_start(canvas, True, True, 5)
b = Gtk.Button('Make Cosine')
b.connect('clicked', _noise, ax, canvas)
box.pack_start(b, False, False, 5)
w.add(box)
w.show_all()
```
